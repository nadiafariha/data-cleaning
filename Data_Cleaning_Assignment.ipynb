```python
import pandas as pd

url = 'https://github.com/nadiafariha/data-cleaning/raw/refs/heads/main/movies_uncleaned.xlsx'
df = pd.read_excel(url)

df
```

```python
# to remove the duplicate row
df = df.drop_duplicates()
# remove those row in which MOVIES column contain Ã anywhere in the string
df = df[~df['MOVIES'].str.contains('Ã')]
# remove STARS column from the table as we dont need it
df = df.drop('STARS', axis=1)
df
```

```python
# YEAR column is super dirty, cleaning it
def clean_year(year_str):
    s = str(year_str).strip()

    if "(" in s:
        s = s[s.index("("):].strip()

    if s.startswith("(") and s.endswith(")"):
        s = s[1:-1].strip()

    s = s.lstrip("-").strip()

    s = s.replace("â€“", "-").replace("–", "-")

    if "-" in s:
        parts = [part.strip() for part in s.split("-")]
        s = "-".join(parts)

    return s



df['YEAR'] = df['YEAR'].apply(clean_year)
df['YEAR'] = df['YEAR'].str.lstrip("(")
df['YEAR'] = df['YEAR'].str.rstrip("-")
df
```

```python
# now removing \n from the GENRE column
df['GENRE'] = df['GENRE'].str.replace('\n', '')
df
```

```python
# make cell empty which have RATING NaN
df['RATING'] = df['RATING'].fillna('')
df
```

```python
#in ONE-LINE column, if there is \n at the start of the string then remove it otherwise we will preserve it
df['ONE-LINE'] = df['ONE-LINE'].str.lstrip('\n')
df
```

```python
#converting VOTES from Double to Int, if there is NaN then make that cell 0
df['VOTES'] = df['VOTES'].fillna(0).astype(int)
# if there is 0 then make it empty cell
df['VOTES'] = df['VOTES'].apply(lambda x: '' if x == 0 else x)
df
```

```python
#converting RunTime from Double to Int, if there is NaN then make that cell 0
df['RunTime'] = df['RunTime'].fillna(0).astype(int)
# if there is 0 then make it empty cell
df['RunTime'] = df['RunTime'].apply(lambda x: '' if x == 0 else x)
df
```

```python
#removing the gross column as we dont need it
df = df.drop('Gross', axis=1)
df
```

```python
# now save the cleaned df as movies_cleaned.xlsx
df.to_excel('movies_cleaned.xlsx', index=False)
```

```python

df['RATING'] = pd.to_numeric(df['RATING'], errors='coerce')

df['GENRE_list'] = df['GENRE'].apply(lambda x: [genre.strip() for genre in x.split(',')])
df_exploded = df.explode('GENRE_list')

pivot_table = df_exploded.pivot_table(
    index='GENRE_list',
    values='RATING',
    aggfunc='mean',
    fill_value=0
)

pivot_table = pivot_table.sort_values(by='RATING', ascending=False)

pivot_table.to_excel('pivot_table_by_genre.xlsx')
```
